<h3> Handling Outliers: </h3>

<p>Dealing with outliers is a vital aspect of data preprocessing, ensuring that exceptionally high or low values do not disproportionately impact the effectiveness of machine learning models.
</p>


<p><b>Common techniques for handling outliers:</b>
 trimming, winsorizing, and transformation.</p>
<p><b>Approach:</b>We utilize the winsorize function from scipy.stats.mstats to substitute extreme values with less extreme ones. The limits parameter indicates the proportion of data at the extremes that should be replaced.</p>

```python
from scipy.stats.mstats import winsorize
# List of columns to handle outliers from training data
train_data_columns_to_handle_outliers = ['TEMP', 'HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT',
       'L1_CLOUD_COVER', 'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER',
       'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 'DEW_POINT', 'MAX_WIND_SPEED',
       'PRECIPITATION', 'PRECIPITATION_TYPE', 'PRESSURE_SEA_LEVEL',
       'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 'AVG_WIND_SPEED',
       'AVG_WIND_DIRECTION', 'DIFFUSE_RADIATION', 'GLOBAL_RADIATION',
       'SUNSHINE_DURATION', 'ZENITH', 'ELEVATION', 'AZIMUTH', 'TIMESTAMP','TARGET']

# Apply winsorizing to handle outliers in the specified columns for train dataset
for column in train_data_columns_to_handle_outliers:
    train_data[column] = winsorize(train_data[column], limits=(0.05, 0.05), inplace=True)
# Display the modified DataFrames
print("Train data after handling outliers:")
print(train_data)
```

<h3> Data Normalization </h3>
<p>
Normalization involves adjusting the values of multiple variables to a comparable range. Common normalization techniques include:

->Scaling the variable so that the average is 0.

->Scaling the variable so that the variance is 1.

->Scaling the variable so that its values range from 0 to 1.

Example:

To illustrate normalization, let's consider scaling the following columns in the training data: "TEMP," "HUMIDITY," "TOTAL_CLOUD_COVER," "L1_CLOUD_HEIGHT," "L1_CLOUD_COVER," "L2_CLOUD_HEIGHT," "L2_CLOUD_COVER," "L3_CLOUD_HEIGHT," "L3_CLOUD_COVER," "DEW_POINT," "MAX_WIND_SPEED," "PRECIPITATION," "PRESSURE_SEA_LEVEL," "PRESSURE_STATION," "VISIBILITY," "AVG_WIND_SPEED," "AVG_WIND_DIRECTION," "DIFFUSE_RADIATION," "GLOBAL_RADIATION," "SUNSHINE_DURATION," "ZENITH," "ELEVATION," "AZIMUTH."

<b>Objective:</b>

Normalize these variables so that their values fall within the range of 0 to 1.

<b>Approach 1:</b>

Replace the original values in the training data with the ratio of each original value to its maximum value.</p>

```python

# List of columns to normalize from the training dataset
train_data_columns_to_normalize = ['TEMP', 'HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT',
       'L1_CLOUD_COVER', 'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER',
       'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 'DEW_POINT', 'MAX_WIND_SPEED',
       'PRECIPITATION', 'PRECIPITATION_TYPE', 'PRESSURE_SEA_LEVEL',
       'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 'AVG_WIND_SPEED',
       'AVG_WIND_DIRECTION', 'DIFFUSE_RADIATION', 'GLOBAL_RADIATION',
       'SUNSHINE_DURATION', 'ZENITH', 'ELEVATION', 'AZIMUTH']

# Normalize the specified columns by dividing by their maximum values
for column in train_data_columns_to_normalize:
    train_data[column] = train_data[column] / train_data[column].max()

# Display the normalized Trained Data DataFrame
print(train_data)
```

```python

# List of columns to normalize from the training dataset
test_data_columns_to_normalize = ['TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT',
       'L1_CLOUD_COVER', 'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER',
       'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 'PRECIPITATION_TYPE']

# Normalize the specified columns by dividing by their maximum values
for column in test_data_columns_to_normalize:
    test_data[column] = test_data[column] / test_data[column].max()

# Display the normalized Trained Data DataFrame
print(test_data)
```

<h3><b>Feature Engineering:</b></h3>

<h4><b>For Training set </b></h4>

```python
#Pressure Difference:
train_data['PRESSURE_DIFFERENCE'] = train_data['PRESSURE_SEA_LEVEL'] - train_data['PRESSURE_STATION']

#Weather-related Features:

# Combine cloud height and cloud cover information
import numpy as np
#Cloud Cover Ratio:
train_data['CLOUD_COVER_RATIO'] = train_data['TOTAL_CLOUD_COVER'] / (train_data['L1_CLOUD_COVER'] + train_data['L2_CLOUD_COVER'] + train_data['L3_CLOUD_COVER'])

# Combine wind speed and direction
train_data['HORIZONTAL_WIND_SPEED'] = train_data['AVG_WIND_SPEED'] * np.cos(np.radians(train_data['AVG_WIND_DIRECTION']))

#calculate wind difference
train_data['WIND_SPEED_DIFFERENCE'] = train_data['MAX_WIND_SPEED'] - train_data['AVG_WIND_SPEED']

# Temperature and Dew Point difference
train_data['TEMP_DEW_DIFF'] = train_data['TEMP'] - train_data['DEW_POINT']

#Solar-related Features:

# Create features related to solar position
train_data['SOLAR_ELEVATION'] = 90 - train_data['ELEVATION']
#train_data['SOLAR_AZIMUTH'] = train_data['AZIMUTH']

# Interaction between solar-related variables
train_data['RADIATION_RATIO'] = train_data['DIFFUSE_RADIATION'] / train_data['GLOBAL_RADIATION']

#Interaction Terms:
train_data['TEMP_HUMIDITY_INTERACTION'] = train_data['TEMP'] * train_data['HUMIDITY']
#visibility to elevation ratio
train_data['VISIBILITY_ELEVATION_RATIO'] = train_data['VISIBILITY'] / train_data['ELEVATION']

#Time-based Aggregations:

# Compute rolling averages
rolling_features = ['TEMP', 'HUMIDITY', 'AVG_WIND_SPEED']
window_size = 12
for feature in rolling_features:
    train_data[f'{feature}_ROLLING_MEAN'] = train_data[feature].rolling(window=window_size).mean()

# Compute cumulative sums
cumulative_features = ['PRECIPITATION', 'SUNSHINE_DURATION']
for feature in cumulative_features:
    train_data[f'{feature}_CUMULATIVE'] = train_data[feature].cumsum()

#Adding new column with categorigal values for the corresponding numeric vales from the PRECIPITATION_TYPE column

# Mapping dictionary for precipitation type descriptions
precipitation_dict = {
    0: 'No precipitation',
    9: 'Type of precipitation unascertainable / incorrect measurement',
    4: 'Type of precipitation unascertainable but depositional precipitation was measured',
    8: 'Liquid and solid precipitation',
    6: 'Liquid precipitation',
    7: 'Solid precipitation'
}

# Column name to insert the new column beside
target_column = 'PRECIPITATION_TYPE'

# Add a new column beside the target column
train_data.insert(train_data.columns.get_loc(target_column) + 1, 'PRECIPITATION_TYPE_DESCRIPTION', train_data[target_column].map(precipitation_dict))

print(train_data.columns,'\n')
print(train_data)
```

# SVM with HyperParameter Tuning

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Specify columns to drop
columns_to_drop = ["PRECIPITATION_TYPE_DESCRIPTION", "TIMESTAMP"]

# Drop specified columns from train data if they exist
X_train = train_data.drop(columns=set(columns_to_drop).intersection(train_data.columns) | {'TARGET'}, axis=1)
y_train = train_data['TARGET']

# Drop specified columns from test data if they exist
X_test = test_data.drop(columns=set(columns_to_drop).intersection(test_data.columns), axis=1)

# Split the train data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Initialize StandardScaler and fit on training data
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_val_scaler = scaler.transform(X_val)
X_test_scaler = scaler.transform(X_test)

# Define the hyperparameter grid
param_grid = {
    'kernel': ['linear', 'rbf'],  # You can add more kernels
    'C': [0.1, 1, 10],  # Adjust the range based on your data
    'gamma': ['scale', 'auto'],  # If using an RBF kernel
}

# Create an SVM regressor
svm_model = SVR()

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=svm_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train_scaler, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train SVM model with the best hyperparameters
best_svm_model = SVR(**best_params)
best_svm_model.fit(X_train_scaler, y_train)

# Make predictions on the validation set
y_val_pred = best_svm_model.predict(X_val_scaler)

# Evaluate the SVM model on the validation set
mae_val = mean_absolute_error(y_val, y_val_pred)
mse_val = mean_squared_error(y_val, y_val_pred)
rmse_val = mean_squared_error(y_val, y_val_pred, squared=False)
r2_val = r2_score(y_val, y_val_pred)

# Display evaluation metrics on the validation set
print(f'Validation Set Metrics:')
print(f'Mean Absolute Error (MAE): {mae_val}')
print(f'Mean Squared Error (MSE): {mse_val}')
print(f'Root Mean Squared Error (RMSE): {rmse_val}')
print(f'R-squared (R2): {r2_val}')

# Make predictions on the test set using the best model
y_pred = best_svm_model.predict(X_test_scaler)

# Ensure y_pred has the correct length
assert len(y_pred) == len(test_data), "Length of y_pred is not equal to the number of rows in test_data"

# Assuming 'ID' is the column name in test_data
id_range = range(1249, 21050)

# Filter and create a DataFrame with 'ID' and 'TARGET'
filtered_predictions = pd.Series(y_pred, name='TARGET')[test_data['ID'].isin(id_range)]

result_df = pd.DataFrame({'ID': test_data['ID'], 'TARGET': filtered_predictions})

# Save the DataFrame to a CSV file
result_df.to_csv('SVM_predictions_with_HyperParametertuning.csv', index=False)
```
