<h4> Handling Outliers </h4>

<b>Why is it necessary?</b>
<p>Handling outliers is an important step in data preprocessing to ensure that extreme values do not unduly influence the performance of machine learning models.
</p>


<p>Common techniques for handling outliers include trimming, winsorizing, and transformation.</p>
<p><b>Approach:</b>We use the winsorize function from scipy.stats.mstats to replace extreme values with less extreme values. The limits parameter specifies the fraction of data at the extremes to replace.</p>

```python
from scipy.stats.mstats import winsorize
# List of columns to handle outliers from training data
train_data_columns_to_handle_outliers = ['TEMP','HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT', 'L1_CLOUD_COVER', 
                    'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER', 'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 
                    'DEW_POINT', 'MAX_WIND_SPEED', 'PRECIPITATION', 'PRECIPITATION_TYPE', 
                    'PRESSURE_SEA_LEVEL', 'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 
                    'AVG_WIND_SPEED', 'AVG_WIND_DIRECTION', "DIFFUSE_RADIATION",'GLOBAL_RADIATION','SUNSHINE_DURATION',
                    'ZENITH','ELEVATION','AZIMUTH','TARGET']

# Apply winsorizing to handle outliers in the specified columns for train dataset
for column in train_data_columns_to_handle_outliers:
    train_data[column] = winsorize(train_data[column], limits=(0.05, 0.05), inplace=True)
# Display the modified DataFrames
print("Train data after handling outliers:")
print(train_data)
```

```python
from scipy.stats.mstats import winsorize
# List of columns to handle outliers from testing data
test_data_columns_to_handle_outliers = ['TEMP','HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT', 'L1_CLOUD_COVER', 
                    'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER', 'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 
                    'DEW_POINT', 'MAX_WIND_SPEED', 'PRECIPITATION', 'PRECIPITATION_TYPE', 
                    'PRESSURE_SEA_LEVEL', 'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 
                    'AVG_WIND_SPEED', 'AVG_WIND_DIRECTION', "DIFFUSE_RADIATION",'GLOBAL_RADIATION','SUNSHINE_DURATION',
                    'ZENITH','ELEVATION','AZIMUTH']

# Apply winsorizing to handle outliers in the specified columns for train dataset
for column in test_data_columns_to_handle_outliers:
    test_data[column] = winsorize(test_data[column], limits=(0.05, 0.05), inplace=True)
# Display the modified DataFrames
print("Test data after handling outliers:")
print(test_data)
```

<h4> Data Normalization </h4>

<b>Why normalization?</b>
<p>Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include 
<ol>
    <li>scaling the variable so the variable average is 0</li>
    <li>scaling the variable so the variance is 1</li> 
    <li>scaling the variable so the variable values range from 0 to 1</li>
</ol>
</p>

<b>Example</b>
<p>To demonstrate normalization, say we want to scale the columns, let's perform normalization on the below columns in the training data:
"TEMP","HUMIDITY","TOTAL_CLOUD_COVER","L1_CLOUD_HEIGHT","L1_CLOUD_COVER","L2_CLOUD_HEIGHT","L2_CLOUD_COVER","L3_CLOUD_HEIGHT",
"L3_CLOUD_COVER","DEW_POINT","MAX_WIND_SPEED","PRECIPITATION","PRESSURE_SEA_LEVEL","PRESSURE_STATION","VISIBILITY",
"AVG_WIND_SPEED","AVG_WIND_DIRECTION","DIFFUSE_RADIATION","GLOBAL_RADIATION","SUNSHINE_DURATION","ZENITH","ELEVATION","AZIMUTH"
.</p>
<p><b>Target:</b> normalize those variables so their value ranges from 0 to 1</p>
<p><b>Approach1:</b> replace the original value by (original value)/(maximum value) in the training data</p>

```python

# List of columns to normalize from the training dataset
train_data_columns_to_normalize = ['TEMP','HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT', 'L1_CLOUD_COVER', 
                    'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER', 'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 
                    'DEW_POINT', 'MAX_WIND_SPEED', 'PRECIPITATION', 'PRECIPITATION_TYPE', 
                    'PRESSURE_SEA_LEVEL', 'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 
                    'AVG_WIND_SPEED', 'AVG_WIND_DIRECTION', "DIFFUSE_RADIATION",'GLOBAL_RADIATION','SUNSHINE_DURATION',
                    'ZENITH','ELEVATION','AZIMUTH','TARGET']

# Normalize the specified columns by dividing by their maximum values
train_data[train_data_columns_to_normalize] = train_data[train_data_columns_to_normalize] / train_data[train_data_columns_to_normalize].max()

# Display the normalized Trained Data DataFrame
print(train_data)
```

```python
# List of columns to normalize from the testing dataset
test_data_columns_to_normalize = ['TEMP','HUMIDITY', 'TOTAL_CLOUD_COVER', 'L1_CLOUD_HEIGHT', 'L1_CLOUD_COVER', 
                    'L2_CLOUD_HEIGHT', 'L2_CLOUD_COVER', 'L3_CLOUD_HEIGHT', 'L3_CLOUD_COVER', 
                    'DEW_POINT', 'MAX_WIND_SPEED', 'PRECIPITATION', 'PRECIPITATION_TYPE', 
                    'PRESSURE_SEA_LEVEL', 'PRESSURE_STATION', 'VISIBILITY', 'WEATHER_TYPE', 
                    'AVG_WIND_SPEED', 'AVG_WIND_DIRECTION', "DIFFUSE_RADIATION",'GLOBAL_RADIATION','SUNSHINE_DURATION',
                    'ZENITH','ELEVATION','AZIMUTH']

# Normalize the specified columns by dividing by their maximum values
test_data[test_data_columns_to_normalize] = test_data[test_data_columns_to_normalize] / test_data[test_data_columns_to_normalize].max()

# Display the normalized Trained Data DataFrame
print(test_data)
```

## Feature Engineering:

<h3>Weather-related Features::</h3>

```python
# Combine cloud height and cloud cover information
import numpy as np
for layer in ['L1', 'L2', 'L3']:
    train_data[f'{layer}_COMBINED'] = train_data[f'{layer}_CLOUD_HEIGHT'] * train_data[f'{layer}_CLOUD_COVER']

# Create a binary variable for precipitation
#train_data['PRECIPITATION_BINARY'] = (train_data['PRECIPITATION'] > 0).astype(int)

# Combine wind speed and direction
train_data['WIND_VECTOR'] = train_data['AVG_WIND_SPEED'] * np.cos(np.radians(train_data['AVG_WIND_DIRECTION']))

# Temperature and Dew Point difference
train_data['TEMP_DEW_DIFF'] = train_data['TEMP'] - train_data['DEW_POINT']
```

<h3>Solar-related Features:</h3>

```python
# Create features related to solar position
train_data['SOLAR_ELEVATION'] = 90 - train_data['ELEVATION']
train_data['SOLAR_AZIMUTH'] = train_data['AZIMUTH']

# Interaction between solar-related variables
train_data['DURATION_PER_RAD'] = train_data['SUNSHINE_DURATION'] / train_data['GLOBAL_RADIATION']
```

<h3>Interaction Terms:</h3>

```python
# Interaction terms
train_data['TEMP_HUMIDITY_INTERACTION'] = train_data['TEMP'] * train_data['HUMIDITY']
train_data['CLOUD_HEIGHT_COVER_INTERACTION'] = train_data['L1_CLOUD_HEIGHT'] * train_data['L1_CLOUD_COVER']
```

<h3>Time-based Aggregations:</h3>

```python
# Compute rolling averages
rolling_features = ['TEMP', 'HUMIDITY', 'AVG_WIND_SPEED']
window_size = 12
for feature in rolling_features:
    train_data[f'{feature}_ROLLING_MEAN'] = train_data[feature].rolling(window=window_size).mean()

# Compute cumulative sums
cumulative_features = ['PRECIPITATION', 'SUNSHINE_DURATION']
for feature in cumulative_features:
    train_data[f'{feature}_CUMULATIVE'] = train_data[feature].cumsum()
```

<p><b>Adding a new column PRECIPITATION_TYPE_DESC with categorigal values for the corresponding numeric vales from the PRECIPITATION_TYPE column </b>
</p>

```python
# Mapping dictionary for precipitation type descriptions
precipitation_mapping = {
    0: 'No precipitation',
    9: 'Type of precipitation unascertainable / incorrect measurement',
    4: 'Type of precipitation unascertainable but depositional precipitation was measured',
    8: 'Liquid and solid precipitation',
    6: 'Liquid precipitation',
    7: 'Solid precipitation'
}

# Column name to insert the new column beside
target_column = 'PRECIPITATION_TYPE'

# Add a new column with corresponding descriptions beside the target column
train_data.insert(train_data.columns.get_loc(target_column) + 1, 'PRECIPITATION_TYPE_DESC', train_data[target_column].map(precipitation_mapping))
# Display the DataFrame
train_data.head(10)
```

```python
# Drop the specified columns from train dataset


train_data = train_data.drop('PRECIPITATION_TYPE_DESC', axis=1)

# Drop 'timestamp' column from both train_data and test_data
train_data = train_data.drop('TIMESTAMP', axis=1)
test_data = test_data.drop('TIMESTAMP', axis=1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



# Specify columns to drop
columns_to_drop = ["PRECIPITATION_TYPE_DESC", "TIMESTAMP"]
# Drop specified columns from train data if they exist
X_train = train_data.drop('TARGET', axis=1)
y_train = train_data['TARGET']



# Split the train_data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Perform polynomial regression with degree=2
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_val_poly = poly.transform(X_val)

model = LinearRegression()
model.fit(X_train_poly, y_train)

# Make predictions on validation set
y_val_pred = model.predict(X_val_poly)

# Evaluate the model on the validation set
mse = mean_squared_error(y_val, y_val_pred)
print(f'Mean Squared Error on validation set: {mse}')

# Now, let's use the trained model to predict on the test_data
X_test_poly = poly.transform(test_data)  # no need to drop columns here
y_test_pred = model.predict(X_test_poly)
print(len(y_test_pred))

assert len(y_test_pred) == len(test_data), "Length of y_pred is not equal to the number of rows in test_data"

# Assuming 'ID' is the column name in test_data
id_range = range(1249, 21050)

# Filter and create a DataFrame with 'ID' and 'TARGET'
filtered_predictions = pd.Series(y_test_pred, name='TARGET')[test_data['ID'].isin(id_range)]
print(filtered_predictions)
result_df = pd.DataFrame({'ID': test_data['ID'], 'TARGET': filtered_predictions})

# Save the DataFrame to a CSV file
result_df.to_csv('PolyNomial_predicted_targets.csv', index=False)
```
